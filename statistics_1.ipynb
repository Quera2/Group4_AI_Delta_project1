```python
#use_shapiro_test_to_show_data_is_not_normal

stat, p_value = shapiro(sample(list(age_401_urban),5000))

print(f"Shapiro-Wilk Test Statistic: {stat}")
print(f"p-value: {p_value}")

alpha = 0.05
if p_value > alpha:
    print("The sample looks Gaussian (fail to reject H0)")
else:
    print("The sample does not look Gaussian (reject H0)")
```

```python
#use_methods_to_normal_data

sample_data = age_401_urban.values.reshape(-1, 1)

yeo_johnson_transformed = power_transform(sample_data, method='yeo-johnson')
yeo_johnson_transformed = yeo_johnson_transformed.flatten()

if np.all(sample_data > 0):
    box_cox_transformed, _ = power_transform(sample_data, method='box-cox', standardize=False)
    box_cox_transformed = box_cox_transformed.flatten()
else:
    box_cox_transformed = None
    print("Box-Cox transformation requires all values to be positive. Skipping Box-Cox.")

stat_original, p_value_original = shapiro(sample_data.flatten())
print(f"Original Data: Shapiro-Wilk Test Statistic: {stat_original}, p-value: {p_value_original}")

stat_yeo, p_value_yeo = shapiro(yeo_johnson_transformed)
print(f"Yeo-Johnson Transformed Data: Shapiro-Wilk Test Statistic: {stat_yeo}, p-value: {p_value_yeo}")

if box_cox_transformed is not None:
    stat_box, p_value_box = shapiro(box_cox_transformed)
    print(f"Box-Cox Transformed Data: Shapiro-Wilk Test Statistic: {stat_box}, p-value: {p_value_box}")

def interpret_shapiro(p_value, alpha=0.05):
    if p_value > alpha:
        return "The sample looks Gaussian (fail to reject H0)"
    else:
        return "The sample does not look Gaussian (reject H0)"

print(f"Original Data: {interpret_shapiro(p_value_original)}")
print(f"Yeo-Johnson Transformed Data: {interpret_shapiro(p_value_yeo)}")
if box_cox_transformed is not None:
    print(f"Box-Cox Transformed Data: {interpret_shapiro(p_value_box)}")
```
