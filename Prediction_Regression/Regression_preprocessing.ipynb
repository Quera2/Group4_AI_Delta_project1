```python
# define functions here
# first one use dummy categorization and make bool column from column given
def dum(dff , col) :
    dp = pd . get_dummies(dff[col], prefix = 'category')
    dff = pd . concat([dff, dp], axis = 1)
    return dff , dp

# second one convert bool value of column into binary
def btb(dff , col) : # bool to binary function
    for item in col:
        dff[item] = dff[item] . astype(int)
    return dff

# dum calculator
def dumd(dff , col) :
    dfff = dum(dff , [col])
    dff = dfff[0]
    dff = btb(dff , dfff[1] . columns)
    dff . drop(columns = col , inplace = True)
    return dff

# MinMaxScaler
def mms(dff , col) : # MinMaxScaler
    mx = MinMaxScaler()
    dff[col] = mx . fit_transform(dff[col])
    return dff
```
