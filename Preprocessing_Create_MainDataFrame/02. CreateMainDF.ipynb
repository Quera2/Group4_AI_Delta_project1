```python
# Library
import pandas as pd
import zipfile
import pickle
import re
import numpy as np
```

### Read Data

```python
#Unzip Data
with zipfile.ZipFile('df_dic.zip', 'r') as myzip:
    myzip.extractall('')


# Open the file in binary mode 
with open('df_dic.pkl', 'rb') as file: 
     
    # Call load method to deserialze 
    df_dic = pickle.load(file) 
```

# Aggrigate All Feature

## Cost

```python
cost_list=[]

for k,sheet in df_dic.items():
    if "P3S" in k: 
        year = int(k.split("P3")[0].replace("R","").replace("U",""))
        if year==99:
            year=1399

        regionType = k[0:1]
        costType = k.split("P3")[1]

        sheet['value'] =  pd.to_numeric(sheet['value'], errors='coerce').fillna(0).astype(np.int64)

        cost = sheet.groupby('Address').agg({'value':'sum'}).rename(columns={"value":'Cost_' + costType})
        cost['year'] = year
        cost['regionType'] = regionType
        
        cost= cost.reset_index().set_index(['Address','year','regionType'])

        cost_list.append(cost)

df_cost = pd.concat(cost_list).groupby(['Address','year','regionType']).agg(sum)

print(df_cost.shape)
df_cost.head()
```

## Detaials of questionnaire

```python
data_list=[]

for k,sheet in df_dic.items():
    if "Data" in k: 
        year = int(k.replace("R","").replace("U","").replace("Data",""))
        if year==99:
            year=1399

        regionType = k[0:1]

        sheet['year'] = year
        sheet['regionType'] = regionType
        
        data= sheet.reset_index().set_index(['Address','year','regionType'])

        data_list.append(data)

df_data = pd.concat(data_list)

print(df_data.shape)
df_data.head()
```

## Habitat Characteristics

```python
habitat_list=[]

for k,sheet in df_dic.items():
    if "P2" in k: 
        year = int(k.replace("R","").replace("U","").replace("P2",""))
        if year==99:
            year=1399

        regionType = k[0:1]

        sheet['year'] = year
        sheet['regionType'] = regionType
        
        habitat= sheet.reset_index().set_index(['Address','year','regionType'])

        habitat_list.append(habitat)

df_habitat = pd.concat(habitat_list)

print(df_habitat.shape)
df_habitat.head()
```

## Social Characteristics

```python
social_list=[]

for k,sheet in df_dic.items():
    if "P1" in k: 
        year = int(k.replace("R","").replace("U","").replace("P1",""))
        if year==99:
            year=1399

        regionType = k[0:1]

        sheet['year'] = year
        sheet['regionType'] = regionType
        
        social= sheet.reset_index().set_index(['Address','year','regionType'])

        social_list.append(social)

df_social_member = pd.concat(social_list)

def age_group(x):
    if x<7:
        return 'baby'
    if 7<= x < 19:
        return 'young'
    if 19<= x < 50:
        return 'middle'
    if 50<= x:
        return 'old'

df_social_member['ageGroup'] = df_social_member.apply(lambda x: age_group(x['age']) , axis=1)

print(df_social_member.shape)
df_social_member.head()
```

```python
# Apply get_dummies to create dummy variables for 'relation','gender','literacy','studying', 'degree','occupationalst', 'maritalst'
df_social_member_with_dummies =pd.get_dummies(df_social_member , columns=['relation' ,'ageGroup' ,'gender','literacy','studying', 'degree','occupationalst', 'maritalst']) 
print(df_social_member_with_dummies.shape)
df_social_member_with_dummies.head()
```

```python
# agg function
columns = df_social_member_with_dummies.columns
dummy_columns = [col for col in columns if "_" in col]
dummy_dic = dict.fromkeys(dummy_columns, 'sum')

df_social = df_social_member_with_dummies.groupby(['Address','year','regionType'])\
                        .agg({'member':'count', 
                              'age':'mean' ,
                               **dummy_dic})
print(df_social.shape)
df_social.head()
```

## Income -> Salary

```python
incomeW_list=[]

for k,sheet in df_dic.items():
    if "P4S01" in k: 
        year = int(k.split("P4")[0].replace("R","").replace("U",""))
        if year==99:
            year=1399

        regionType = k[0:1]
        incomeType = k.split("P4")[1]

        sheet['netincome_w_y'] =  pd.to_numeric(sheet['netincome_w_y'], errors='coerce').fillna(0).astype(np.int64)

        sheet_withdummy = pd.get_dummies(sheet , columns=['employed_w' , 'status_w'])


        incomeW = sheet_withdummy.groupby('Address').agg({ 'member':'nunique',
                                                'employed_w_1':'sum','employed_w_2':'sum',
                                                'status_w_1':'sum','status_w_2':'sum','status_w_3':'sum',
                                                'days_w':'sum', 
                                                'netincome_w_y':'sum'}).rename(columns={'member':'member_incomeW'})
        incomeW['year'] = year
        incomeW['regionType'] = regionType
        
        incomeW= incomeW.reset_index().set_index(['Address','year','regionType'])

        incomeW_list.append(incomeW)

df_incomeW = pd.concat(incomeW_list).groupby(['Address','year','regionType']).agg(sum)

print(df_incomeW.shape)
df_incomeW.head()
```

## Income -> Freelance

```python
incomeS_list=[]

for k,sheet in df_dic.items():
    if "P4S02" in k: 
        year = int(k.split("P4")[0].replace("R","").replace("U",""))
        if year==99:
            year=1399

        regionType = k[0:1]
        incomeType = k.split("P4")[1]

        sheet['income_s_y'] =  pd.to_numeric(sheet['income_s_y'], errors='coerce').fillna(0).astype(np.int64)
        sheet['employed_s'] = pd.to_numeric(sheet['employed_s'], errors='coerce').fillna(1).astype(np.int64)
        sheet['status_s'] = pd.to_numeric(sheet['status_s'], errors='coerce').fillna(1).astype(np.int64)
        sheet['agriculture'] = pd.to_numeric(sheet['agriculture'], errors='coerce').fillna(1).astype(np.int64)

        sheet_withdummy = pd.get_dummies(sheet , columns=['employed_s' , 'status_s', 'agriculture'])


        incomeS = sheet_withdummy.groupby('Address').agg({ 'member':'nunique',
                                                'employed_s_1':'sum','employed_s_2':'sum',
                                                'status_s_4':'sum','status_s_5':'sum','status_s_6':'sum',
                                                'agriculture_1':"sum",'agriculture_2':"sum",
                                                'days_s':'sum', 
                                                'income_s_y':'sum'}).rename(columns={'member':'member_incomeS'})
        incomeS['year'] = year
        incomeS['regionType'] = regionType
        
        incomeS= incomeS.reset_index().set_index(['Address','year','regionType'])

        incomeS_list.append(incomeS)

df_incomeS = pd.concat(incomeS_list).groupby(['Address','year','regionType']).agg(sum)

print(df_incomeS.shape)
df_incomeS.head()
```

## Income -> other

```python
incomeO_list=[]

for k,sheet in df_dic.items():
    if "P4S03" in k: 
        year = int(k.split("P4")[0].replace("R","").replace("U",""))
        if year==99:
            year=1399

        regionType = k[0:1]
        incomeType = k.split("P4")[1]


        for col in sheet.columns:
            sheet[col] =  pd.to_numeric(sheet[col], errors='coerce').fillna(0).astype(np.int64)


        sheet['income_o_y'] = sheet['income_pension'] + sheet['income_rent'] + sheet['income_interest'] \
                            + sheet['income_aid'] + sheet['income_resale'] + sheet['income_transfer']


        incomeO = sheet.groupby('Address').agg({ 'member':'nunique',
                                                'income_o_y':'sum'}).rename(columns={'member':'member_incomeO'})
        incomeO['year'] = year
        incomeO['regionType'] = regionType
        
        incomeO= incomeO.reset_index().set_index(['Address','year','regionType'])

        incomeO_list.append(incomeO)

df_incomeO = pd.concat(incomeO_list).groupby(['Address','year','regionType']).agg(sum)

print(df_incomeO.shape)
df_incomeO.head()
```

## Income -> Subcidy

```python
incomeSu_list=[]

for k,sheet in df_dic.items():
    if "P4S04" in k: 
        year = int(k.split("P4")[0].replace("R","").replace("U",""))
        if year==99:
            year=1399

        regionType = k[0:1]
        incomeType = k.split("P4")[1]


        for col in sheet.columns:
            sheet[col] =  pd.to_numeric(sheet[col], errors='coerce').fillna(0).astype(np.int64)


        incomeSu = sheet.groupby('Address').agg({ 'member':'nunique',
                                                'subsidy':'sum'}).rename(columns={'member':'member_incomeSu',
                                                                                  'subsidy':'income_Su_y'})
        incomeSu['year'] = year
        incomeSu['regionType'] = regionType
        
        incomeSu= incomeSu.reset_index().set_index(['Address','year','regionType'])

        incomeSu_list.append(incomeSu)

df_incomeSu = pd.concat(incomeSu_list).groupby(['Address','year','regionType']).agg(sum)

print(df_incomeSu.shape)
df_incomeSu.head()
```

# Aggriage All DF

```python
df_all = { 'Costs':df_cost,
            'Details':df_data,
            'Habitat Charachteristics':df_habitat,
            'Social Charachteristics':df_social,
            'Salary income':df_incomeW,
            'Freelance income':df_incomeS,
            'Other income':df_incomeO,
            'Subcidy income':df_incomeSu}

for k,d in df_all.items():
    print(k, "--->", d.shape)
```

```python
df_main = pd.concat([df for k,df in df_all.items()] ).groupby(['Address','year','regionType']).agg(sum)
print('df_main shape:' ,df_main.shape)
df_main.head()
```
